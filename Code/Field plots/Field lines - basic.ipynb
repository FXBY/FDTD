# Plotting field lines with `matplotlib` in 2D
- like with plotting vector fields it is easy to plot field lines of vector fields, but the plots are rather basic and sketchy (recall that field lines do not start and end in the middle of nowhere as you can see in the plots!)
- **Warning:** Do not plot field lines in A Level problems as you see in these plots. You have to be much more pedantic than this to get the marks! 
- for better and more sophisticated field line plots you have to solve the corresponding ODE of the field line and plot it. This can be done using `scipy.solve_ivp` or `scipy.ode`. (Do not use the Euler method for this, as this is too unreliable.)
- when plotting field lines in 3D you might want to consider using the library `plotly` rather than `matplotlib`

```python
# Importing packages
import numpy as np
import matplotlib.pyplot as plt

# This is a magic command for plots being plotted inline in the notebook. 
%matplotlib inline
```

### Basic 2D field line plot
- the process is the same as when we wish to plot a vector field

Also try the vector fields:

$$
    V(x,y) = (-y,x)\qquad V(x,y) = (-y,-x)
$$

See how a simple change in sign can change the geometry of the vector field completely!

```python
# The vector field function
def V(x:float, y:float)-> tuple:
    return (-x,-y)

# Create the grid to calculate and display the vector field on 
RangeX = np.arange(-10, 10, 1)
RangeY = np.arange(-10, 10, 1)
X, Y = np.meshgrid(RangeX, RangeY)

# Calculate the Vector field at each point on the grid by applying V to each component of the array X and Y.
# As the plotting functions of matplotlib want separate x, and y-components we use tuple unpacking
V_X, V_Y = V(X,Y) 

# We create a figure here, but we could have used `plt.streamplot` directly as well
fig, ax = plt.subplots(figsize = (10,8))

# Sketch field lines
#ax.streamplot(X, Y, V_X, V_Y)
ax.quiver(X, Y, V_X, V_Y)

plt.show()
```

### Adding colours to display magnitude in 2D

```python
# The vector field function
def V(x:float, y:float)-> tuple:
    return (-x,-y)

# Create the grid to calculate and display the vector field on 
RangeX = np.arange(-10, 10, 1)
RangeY = np.arange(-10, 10, 1)
X, Y = np.meshgrid(RangeX, RangeY)

# Calculate the Vector field at each point on the grid by applying V to each component of the array X and Y.
# As the plotting functions of matplotlib want separate x, and y-components we use tuple unpacking
V_X, V_Y = V(X,Y) 

# Calculate magnitude for normalising and colouring
Mag = np.hypot(V_X, V_Y)

fig, ax = plt.subplots(figsize = (10,8))

ax.streamplot(X, Y, V_X, V_Y, color = Mag, cmap='plasma')

plt.show()
```
